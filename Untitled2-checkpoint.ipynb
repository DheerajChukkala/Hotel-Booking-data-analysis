# Data Analysis and Visualizations

```python
cancelled_perc = dataset['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
plt.figure(figsize=(5,4))
plt.title('Count of Reservation status ')
plt.bar(['Not cancelled','Cancelled'],dataset['is_canceled'].value_counts(),edgecolor='r',width=0.5)
```

```python
plt.figure(figsize=(8,6))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=dataset,palette='inferno')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status of different hotels',size=20)
plt.xlabel('Hotel')
plt.ylabel('No of Reservations')
plt.legend(['Not cancelled','Cancelled'])
plt.show()
#The get_legend_handles_labels() function retrieves the handles and labels of the legend items from the ax1 object. 
#The handles represent the bars or patches in the plot, and the labels correspond to the legend labels.
#The legend() function is then called on ax1 with the bbox_to_anchor parameter set to (1, 1).
#This parameter specifies the coordinates of the legend box's anchor point,which in this case isthetop-right corner of the plot.
```

```python
Resort_Hotel=dataset[dataset['hotel'] =='Resort Hotel']
Resort_Hotel['is_canceled'].value_counts(normalize=True)
```

```python
City_Hotel=dataset[dataset['hotel'] =='City Hotel']
City_Hotel['is_canceled'].value_counts(normalize=True)
```

```python
print(Resort_Hotel['adr'].dtypes)
print(City_Hotel['adr'].dtypes)
```

```python
Resort_Hotel=Resort_Hotel.groupby('reservation_status_date')[['adr']].mean()
City_Hotel=City_Hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in city hotel and resort hotel',fontsize=30)
plt.plot(Resort_Hotel.index,Resort_Hotel['adr'],label='Resort Hotel')
plt.plot(City_Hotel.index,City_Hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
dataset['Month']=dataset['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='Month',hue='is_canceled',data=dataset,palette='bright')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per Month',size=20)
plt.xlabel('Month')
plt.ylabel('No of Reservations')
plt.legend(['Not cancelled','Cancelled'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=20)
sns.barplot('Month','adr',data=dataset[dataset['is_canceled'] == 1].groupby('Month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancelled_data= dataset[dataset['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(10,10))
plt.title('Top 10 countries with reservations cancelled')
plt.pie(top_10_country, autopct = '%.2f',labels=top_10_country.index)
plt.show()
```
